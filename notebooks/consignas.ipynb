```python
#Agregamos la columna edad
from datetime import datetime, date

def calculate_age(born):
    born = datetime.strptime(born, '%Y-%m-%d')
    today = date.today()
    return today.year - born.year - ((today.month, today.day) < (born.month, born.day))

df.loc[:, 'edad'] = df[df[nacimiento].notnull()].fecha_nacimiento.apply(calculate_age)
```

```python
#Usamos columnas seleccionadas
compras= "monto_compra_movimiento"
cuota= "importe_cuota_movimiento"
productos= "producto_movimiento"
moneda= "moneda_movimiento"
fecha_nac= "fecha_nacimiento"
sexo="sexo"
provincia="provincia"
antig_cuenta="antiguedad_de_la_cuenta"
fecha= "fecha_cupon_movimiento"
edad="edad"
df2 = df[[
    compras,
    cuota,
    productos,
    moneda,
    fecha_nac,
    sexo,
    provincia,
    antig_cuenta,
    fecha,
    edad
]]
#Convertimos la columna de fecha a datetime
df2.loc[:, fecha] = df2[fecha].apply(lambda fecha: datetime.strptime(fecha, '%Y-%m-%d'))
#Creamos una nueva columna para mostrar el año y la fecha en una nueva columna (para poder hacer el merge con la tabla inflacionaria)
df2.loc[:, 'mes'] = df2[fecha].apply(lambda fecha: datetime.strftime(fecha, '%Y-%m'))
df2
```

```python
from datetime import datetime, date
data_inflacion_por_mes = [
    ['2020-01',1.504],
    ['2020-02',1.474],
    ['2020-03',1.426],
    ['2020-04',1.405],
    ['2020-05',1.384],
    ['2020-06',1.354],
    ['2020-07',1.328],
    ['2020-08',1.293],
    ['2020-09',1.257],
    ['2020-10',1.212],
    ['2020-11',1.175],
    ['2020-12',1.130],
    ['2021-01',1.086],
    ['2021-02',1.048],
    ['2021-03',1.000]
]

df_de_inflacion = pd.DataFrame(data_inflacion_por_mes, columns = ['mes', 'multiplicador'])
```

```python

df3 = df2.merge(df_de_inflacion, on='mes')

def normalizar_monto(monto, multiplicador):
    #Acá deberíamos pasar también la moneda y la cotización
    return monto * multiplicador

df3.loc[:, 'monto_normalizado'] = df3[[compras, 'multiplicador']].apply(lambda row: normalizar_monto(row[compras], row['multiplicador']), axis=1)
df3.head()
```
